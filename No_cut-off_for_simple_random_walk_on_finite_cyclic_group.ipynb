# No cut-off for simple random walk on finite cyclic group

For certain families of random walks on finite groups, it is possible to observe a 'cut-off'. It describes a sharp transition in the convergence of finite ergodic Markov chains to stationarity. For example, this cut-off phenomenon occurs when one tries to shuffle a deck of cards with the 'Riffle shuffle'. 

This notebook aims at giving a simple example of random walk on finite group without cut-off. We consider the (additive) cyclic group of order $d \ge 7$, $d$ odd, and the following random walk: from a state $k$ (mod $d$), the following state is either $k+1$ (mod $d$) or $k-1$ (mod $d$) with probability $\frac{1}{2}$.
The random walk starts in $0$.

It is possible to prove that there is not cut-off for this random walk, using representation theory (see the handout).


One can see that there is no cut-off by plotting the evolution of the total variation distance between $Q^{*k}$: the distribution of the $k$-th state $(k = 0, ..., n-1)$ of the random walk, and $U$: the uniform distribution on $\mathbb{Z}/d\mathbb{Z}$.

```python
from simple_RW_on_cyclic_group import TV_plot
```

Here is an example with $\mathbb{Z}/51\mathbb{Z}$:

```python
TV_plot(1500, 51)
```

Let's see what happen when $d$ is even, for example with $\mathbb{Z}/52\mathbb{Z}$:

```python
TV_plot(600, 52)
```

When $d$ is even, it is easy to see that the transition matrix of the Markov chain isn't regular, due to the parity of $d$. Therefore, we can't apply the Perron-Froebenius theorem to prove the convergence:

\begin{equation}
\forall n \in \mathbb{N}, \quad || Q^{*n} - U ||_{TV} \le \sqrt{d}|(1-\delta)^n
\end{equation}

Where $\delta$ is the spectral gap of the transition matrix.
